```python
import pandas as pd
import numpy as np
```

```python
# Дані про позашляховики
data = {
    'Brand': ['Lexus', 'BMW', 'Audi', 'Mercedes-Benz', 'Volvo', 'Kia', 'Porsche', 'Land Rover', 'Tesla', 'Jaguar'],
    'Price': [35000, 45000, 28000, 37000, 40000, 25000, 42000, 32000, 50000, 46000],
    'SafetyRating': [5, 3, 4, 4, 5, 4, 5, 5, 4, 5],
    'CargoSpace': [520, 400, 470, 530, 550, 440, 510, 460, 600, 480],
    'Horsepower': [210, 280, 170, 240, 300, 190, 250, 220, 320, 270],
    'FuelEconomy': [8.0, 11.2, 7.3, 12.0, 10.1, 7.9, 9.8, 8.5, 12.5, 9.0],
    'WarrantyYears': [5, 2, 4, 3, 6, 7, 5, 4, 6, 4],
    'GroundClearance': [210, 205, 215, 198, 202, 220, 235, 195, 245, 200]
}

# Створення DataFrame
suv_df = pd.DataFrame(data)
suv_df
```

```python
# Зміна індексу на назву брендів
suv_df = suv_df.set_index('Brand')
suv_df.index.name = None

# Транспонування таблиці
suv_df = suv_df.T

# Додаємо рядок з критеріями (min або max)
suv_df['Criterion'] = ['min', 'max', 'max', 'max', 'min', 'max', 'max']
suv_df
```

```python
# Збереження в CSV
suv_df.to_csv('data.csv')
```

```python
# Читання з CSV файлу
suv_df = pd.read_csv('data.csv', index_col=0)
suv_df
```

```python
# Нормалізація за критеріями
def normalize_criterion(row):
    if row['Criterion'] == 'min':
        return row[:-1] / row[:-1].sum()  # Нормалізація для мінімізаційних критеріїв
    return (1 / row[:-1]) / sum(1 / row[:-1])  # Нормалізація для максимізаційних критеріїв

# Обчислення зваженої оцінки
def calculate_efficiency(row, weights):
    return (1 - row) ** (-1) @ weights
```

```python
# Застосовуємо нормалізацію
suv_df = suv_df.apply(normalize_criterion, axis=1)
suv_df = suv_df.T
suv_df
```

```python
# Встановлюємо ваги
weights = np.array([0.1, 0.3, 0.1, 0.15, 0.1, 0.05, 0.1, 0.1])
weights = weights / sum(weights)  # Нормалізація ваг

# Обчислюємо загальний бал
suv_df['EfficiencyScore'] = suv_df.apply(lambda x: calculate_efficiency(x, weights), axis=1)

# Сортуємо за отриманими оцінками
suv_df = suv_df.sort_values(by='EfficiencyScore')
suv_df['EfficiencyScore']
```

```python
# Виводимо результат
print(f'Best SUV Brand: {suv_df.iloc[0].name}')
print(f'Integrated Efficiency Score: {suv_df.iloc[0, -1]}')
```
